# FACE DETECTION

```python
face_cascade=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
def detect_face(img):

    face_img=img.copy()

    face_rects = face_cascade.detectMultiScale(face_img)


    for(x,y,w,h) in face_rects:
        cv2.rectangle(face_img, (x,y), (x+w, y+h), (255,255,255), 10)
        cv2.imshow("Face Detection", face_img)
    return face_img
```

```python
result1= detect_face(with_out_glass2)
plt.imshow(result1, cmap='gray')
plt.show()
```

```python
result2= detect_face(with_out_glass1)
plt.imshow(result2, cmap='gray')
plt.show()
```

```python
result3= detect_face(grp_pic)
plt.imshow(result3, cmap='gray')
plt.show()
```

# EYE DETECTION

```python
eye_cascade=cv2.CascadeClassifier('haarcascade_eye.xml')
```

```python
def detect_eyes(img):
    eye_img = img.copy()
    eye_rect = eye_cascade.detectMultiScale(eye_img, scaleFactor=1.2, minNeighbors=5)

    for (x, y, w, h) in eye_rect:
        cv2.rectangle(eye_img, (x, y), (x + w, y + h), (255, 255, 255), 10)
        
    return eye_img
```

```python
result4=detect_eyes(with_out_glass2)
plt.imshow(result4, cmap='gray')
plt.show()
```

```python
result5=detect_eyes(with_out_glass1)
plt.imshow(result5, cmap='gray')
plt.show()
```

```python
result6=detect_eyes(grp_pic)
plt.imshow(result6, cmap='gray')
plt.show()
```

# Conjunction With Video

```python
cap = cv2.VideoCapture(0)

plt.ion()
fig, ex=plt.subplots()

ret , frame = cap.read(0)
frame=detect_face(frame)
im = ex.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
plt.title('Video face Detection')

while True:
    ret, frame = cap.read(0)

    frame= detect_face(frame)

    im.set_data(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
    plt.pause(0.10)

cap.release()
plt.clsoe()
```

```python
cap = cv2.VideoCapture(0)

plt.ion()
fig, ex=plt.subplots()

ret , frame = cap.read(0)
frame=detect_eyes(frame)
im = ex.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
plt.title('Video eye Detection')

while True:
    ret, frame = cap.read(0)

    frame= detect_eyes(frame)

    im.set_data(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
    plt.pause(0.10)

cap.release()
plt.clsoe()
```
